```python
import re
import unicodedata



def normalizar_instituicao(nome):

    mapeamento = [
        (r"(FLUMIN|UFF|INFES/UFF|IC/UFF|PPGCI/UFF)",
         "Universidade Federal Fluminense"),

        (r"(SAO PAULO|USP|ICMC|POLITECNICA|EACH|UNIVERSITY OF SÃO PAULO)",
         "Universidade de São Paulo"),

        (r"(MINAS GERAIS|UFMG|BIOINFORMATICA|DCC/UFMG)",
         "Universidade Federal de Minas Gerais"),

        (r"(RIO DE JANEIRO|UFRJ|COPPE|PPGI/UFRJ|BNDES|IBCCF|PPGEN|CISI/COPPE)",
         "Universidade Federal do Rio de Janeiro"),

        (r"(CAMPINA GRANDE|UFCG|CAMINA GRANDE)",
         "Universidade Federal de Campina Grande"),

        (r"(CAMPINAS|UNICAMP|CEPAGRI)",
         "Universidade Estadual de Campinas"),

        (r"(CEARA|UFC)",
         "Universidade Federal do Ceará"),

        (r"(PERNAMBUCO|UFPE|CIN/UFPE|UFRPE|AGRESTE)",
         "Universidade Federal de Pernambuco"),

        (r"(SAO CARLOS|UFSCAR)",
         "Universidade Federal de São Carlos"),

        (r"(PUC RIO|PONTIFICAL CATHOLIC UNIVERSITY OF RIO DE JANEIRO)",
         "Pontifícia Universidade Católica do Rio de Janeiro"),

        (r"(GOIAS|UFG)",
         "Universidade Federal de Goiás"),

        (r"(PARANA|UFPR)",
         "Universidade Federal do Paraná"),

        (r"(UBERLANDIA|UFU)",
         "Universidade Federal de Uberlândia"),

        (r"(OURO PRETO|UFOP)",
         "Universidade Federal de Ouro Preto"),

        (r"(SANTA CATARINA|UFSC)",
         "Universidade Federal de Santa Catarina"),

        (r"(RIO GRANDE DO SUL|UFRGS)",
         "Universidade Federal do Rio Grande do Sul"),

        (r"(AMAZONAS|UFAM)",
         "Universidade Federal do Amazonas"),

        (r"(BRASILIA|UNB)",
         "Universidade de Brasília"),

        (r"(SAO JOAO DEL REI|UFSJ)",
         "Universidade Federal de São João Del Rei"),

        (r"(FORTALEZA|UNIFOR)",
         "Universidade de Fortaleza"),

        (r"(INSTITUTO FEDERAL|IFSP|IFMG|IFPR|IFNMG|IFPB|IFRS|IFSUL|IF SUDOESTE)",
         "Institutos Federais"),

        (r"(CEFET)",
         "Centro Federal de Educação Tecnológica"),

        (r"(UTFPR|TECHNOLOGY PARANA)",
         "Universidade Tecnológica Federal do Paraná"),

        (r"(AERONAUTICA|ITA)",
         "Instituto Tecnológico de Aeronáutica"),

        (r"(FIOCRUZ|OSWALDO CRUZ)",
         "Fundação Oswaldo Cruz"),

        (r"(EMBRAPA)",
         "Empresa Brasileira de Pesquisa Agropecuária"),

        (r"(IBM)",
         "IBM Research"),

        (r"(JUSBRASIL)",
         "Jusbrasil"),

        (r"(NO AFFILIATION DECLARED	|NOT INFORMED)",
         "Sem Afiliação"),

        (r"(BRAZIL)",
         "BRASIL")
    ]

    if pd.isna(nome):
        return None

    for padrao, nome_padrao in mapeamento:
        if re.search(padrao, nome):
            return nome_padrao.upper()

    return nome
```

```python
df['autores'] = autores.groupby(level=0).agg({'nome': lambda x: ', '.join(x)})

# Separar os autores em colunas distintas
df['autor_1'] = df['autores'].str.split(',').str[0].str.strip().str.upper().fillna(None)
df['autor_2'] = df['autores'].str.split(',').str[1].str.strip().str.upper().fillna(None)
df['autor_3'] = df['autores'].str.split(',').str[2].str.strip().str.upper().fillna(None)
df['autor_4'] = df['autores'].str.split(',').str[3].str.strip().str.upper().fillna(None)
df['autor_5'] = df['autores'].str.split(',').str[4].str.strip().str.upper().fillna(None)
df['autor_6'] = df['autores'].str.split(',').str[5].str.strip().str.upper().fillna(None)
df['autor_7'] = df['autores'].str.split(',').str[6].str.strip().str.upper().fillna(None)
df['autor_8'] = df['autores'].str.split(',').str[7].str.strip().str.upper().fillna(None)
```

Criar ID 

```python
df['id'] = df['url'].str.split('/').str[-1].str.replace('.html', '', regex=False)
```

```python
# Reordenar as colunas para colocar o ID no início
colunas = ['id'] + [col for col in df.columns if col != 'id']
df = df[colunas]
```

```python
df = df.drop(columns=['autores', 'url', 'doi', 'instituicoes']).reset_index(drop=True)
```

Keywords

```python
keywords = df.explode('keywords')

df["keywords"] = (
    keywords.groupby(level=0)
            .agg({'keywords': lambda x: ', '.join(x.dropna().astype(str))})
)


# Separar as palavras-chave em colunas distintas
df['keyword_1'] = df['keywords'].str.split(',').str[0].str.strip().str.upper().fillna(None)
df['keyword_2'] = df['keywords'].str.split(',').str[1].str.strip().str.upper().fillna(None)
df['keyword_3'] = df['keywords'].str.split(',').str[2].str.strip().str.upper().fillna(None)
df['keyword_4'] = df['keywords'].str.split(',').str[3].str.strip().str.upper().fillna(None)
df['keyword_5'] = df['keywords'].str.split(',').str[4].str.strip().str.upper().fillna(None)
```
