```python
import re
import unicodedata



def normalizar_instituicao(nome):

    mapeamento = [
        (r"(FLUMIN|UFF|INFES/UFF|IC/UFF|PPGCI/UFF)",
         "Universidade Federal Fluminense"),

        (r"(SAO PAULO|USP|ICMC|POLITECNICA|EACH|UNIVERSITY OF SÃO PAULO)",
         "Universidade de São Paulo"),

        (r"(MINAS GERAIS|UFMG|BIOINFORMATICA|DCC/UFMG)",
         "Universidade Federal de Minas Gerais"),

        (r"(RIO DE JANEIRO|UFRJ|COPPE|PPGI/UFRJ|BNDES|IBCCF|PPGEN|CISI/COPPE)",
         "Universidade Federal do Rio de Janeiro"),

        (r"(CAMPINA GRANDE|UFCG|CAMINA GRANDE)",
         "Universidade Federal de Campina Grande"),

        (r"(CAMPINAS|UNICAMP|CEPAGRI)",
         "Universidade Estadual de Campinas"),

        (r"(CEARA|UFC)",
         "Universidade Federal do Ceará"),

        (r"(PERNAMBUCO|UFPE|CIN/UFPE|UFRPE|AGRESTE)",
         "Universidade Federal de Pernambuco"),

        (r"(SAO CARLOS|UFSCAR)",
         "Universidade Federal de São Carlos"),

        (r"(PUC RIO|PONTIFICAL CATHOLIC UNIVERSITY OF RIO DE JANEIRO)",
         "Pontifícia Universidade Católica do Rio de Janeiro"),

        (r"(GOIAS|UFG)",
         "Universidade Federal de Goiás"),

        (r"(PARANA|UFPR)",
         "Universidade Federal do Paraná"),

        (r"(UBERLANDIA|UFU)",
         "Universidade Federal de Uberlândia"),

        (r"(OURO PRETO|UFOP)",
         "Universidade Federal de Ouro Preto"),

        (r"(SANTA CATARINA|UFSC)",
         "Universidade Federal de Santa Catarina"),

        (r"(RIO GRANDE DO SUL|UFRGS)",
         "Universidade Federal do Rio Grande do Sul"),

        (r"(AMAZONAS|UFAM)",
         "Universidade Federal do Amazonas"),

        (r"(BRASILIA|UNB)",
         "Universidade de Brasília"),

        (r"(SAO JOAO DEL REI|UFSJ)",
         "Universidade Federal de São João Del Rei"),

        (r"(FORTALEZA|UNIFOR)",
         "Universidade de Fortaleza"),

        (r"(INSTITUTO FEDERAL|IFSP|IFMG|IFPR|IFNMG|IFPB|IFRS|IFSUL|IF SUDOESTE)",
         "Institutos Federais"),

        (r"(CEFET)",
         "Centro Federal de Educação Tecnológica"),

        (r"(UTFPR|TECHNOLOGY PARANA)",
         "Universidade Tecnológica Federal do Paraná"),

        (r"(AERONAUTICA|ITA)",
         "Instituto Tecnológico de Aeronáutica"),

        (r"(FIOCRUZ|OSWALDO CRUZ)",
         "Fundação Oswaldo Cruz"),

        (r"(EMBRAPA)",
         "Empresa Brasileira de Pesquisa Agropecuária"),

        (r"(IBM)",
         "IBM Research"),

        (r"(JUSBRASIL)",
         "Jusbrasil"),

        (r"(NO AFFILIATION DECLARED	|NOT INFORMED)",
         "Sem Afiliação"),

        (r"(BRAZIL)",
         "BRASIL")
    ]

    if pd.isna(nome):
        return None

    for padrao, nome_padrao in mapeamento:
        if re.search(padrao, nome):
            return nome_padrao.upper()

    return nome
```

```python
#Melhor ate o momento

import pandas as pd
import difflib
import re

def normalizar_para_comparacao(texto):
    # Remove pontos, acentos e espaços extras para facilitar a comparação
    texto = re.sub(r'\.', ' ', texto.upper())
    return " ".join(texto.split())

def sao_mesma_pessoa(nome1, nome2, threshold=0.8):
    # 1. Teste de similaridade padrão (para erros de digitação)
    ratio = difflib.SequenceMatcher(None, nome1, nome2).ratio()
    if ratio >= threshold:
        return True
    
    # 2. Teste de "Contenção" (para nomes abreviados ou curtos)
    # Ex: "AGMA TRAINA" dentro de "AGMA JUCI MACHADO TRAINA"
    n1 = normalizar_para_comparacao(nome1).split()
    n2 = normalizar_para_comparacao(nome2).split()
    
    # Identifica qual é o nome curto e qual é o longo
    curto, longo = (n1, n2) if len(n1) < len(n2) else (n2, n1)
    
    # Se o nome curto tem pelo menos 2 termos (Primeiro e Último nome)
    if len(curto) >= 2:
        # Verifica se o primeiro e o último nome batem exatamente
        if curto[0] == longo[0] and curto[-1] == longo[-1]:
            # Verifica se os outros termos do nome curto são iniciais ou partes do longo
            match_count = 0
            for termo in curto:
                if any(termo == parte or (len(termo) == 1 and parte.startswith(termo)) for parte in longo):
                    match_count += 1
            if match_count == len(curto):
                return True
                
    return False

def criar_mapeamento_fuzzy_avancado(lista_nomes):
    mapping = {}
    # Ordenamos do maior para o menor para que o nome completo seja o "pai"
    lista_ordenada = sorted(list(set(lista_nomes)), key=len, reverse=True)
    processed = set()

    for i in range(len(lista_ordenada)):
        nome_longo = lista_ordenada[i]
        if nome_longo in processed:
            continue
        
        for j in range(i + 1, len(lista_ordenada)):
            nome_curto = lista_ordenada[j]
            if nome_curto in processed:
                continue
            
            if sao_mesma_pessoa(nome_longo, nome_curto):
                mapping[nome_curto] = nome_longo
                processed.add(nome_curto)
        
        processed.add(nome_longo)
    return mapping

# --- APLICANDO NO SEU FLUXO ---

cols_autores = [f'autor_{i}' for i in range(1, 9)]
todos_nomes = pd.Series(df[cols_autores].values.ravel()).dropna().unique()

# Gera o mapeamento com a nova lógica
mapeamento_limpo = criar_mapeamento_fuzzy_avancado(todos_nomes)

# Aplica a substituição
df[cols_autores] = df[cols_autores].replace(mapeamento_limpo)

# Gera a lista final
nomes_pessoas_lista = sorted(pd.Series(df[cols_autores].values.ravel()).dropna().unique().tolist())
```

```python
import pandas as pd
import re
import unicodedata
import difflib

def remover_acentos_e_limpar(texto):
    if not texto: return ""
    nfkd = unicodedata.normalize('NFD', str(texto).upper())
    texto_sem_acento = "".join([c for c in nfkd if unicodedata.category(c) != 'Mn'])
    return re.sub(r'[^A-Z\s]', ' ', texto_sem_acento)

def normalizar_para_comparacao(texto):
    limpo = remover_acentos_e_limpar(texto)
    return " ".join(limpo.split())

def sao_mesma_pessoa(nome1, nome2, threshold=0.85):
    n1_str = normalizar_para_comparacao(nome1)
    n2_str = normalizar_para_comparacao(nome2)
    
    if difflib.SequenceMatcher(None, n1_str, n2_str).ratio() >= threshold:
        return True
    
    t1, t2 = n1_str.split(), n2_str.split()
    curto, longo = (t1, t2) if len(t1) < len(t2) else (t2, t1)
    
    if len(curto) >= 2:
        if curto[0] == longo[0] and curto[-1] == longo[-1]:
            match_count = 0
            for termo in curto:
                if any(termo == parte or (len(termo) == 1 and parte.startswith(termo)) for parte in longo):
                    match_count += 1
            if match_count == len(curto):
                return True
    return False

def criar_mapeamento_fuzzy_total(lista_nomes):
    mapping = {}
    lista_ordenada = sorted(list(set(lista_nomes)), key=len, reverse=True)
    processed = set()

    for i in range(len(lista_ordenada)):
        nome_longo = lista_ordenada[i]
        if nome_longo in processed: continue
        
        for j in range(i + 1, len(lista_ordenada)):
            nome_curto = lista_ordenada[j]
            if nome_curto in processed: continue
            
            if sao_mesma_pessoa(nome_longo, nome_curto):
                mapping[nome_curto] = nome_longo
                processed.add(nome_curto)
        processed.add(nome_longo)
    return mapping

cols_autores = [f'autor_{i}' for i in range(1, 9)]

todos_nomes = pd.Series(df[cols_autores].values.ravel()).dropna().unique()

mapeamento_limpo = criar_mapeamento_fuzzy_total(todos_nomes)

df[cols_autores] = df[cols_autores].replace(mapeamento_limpo)

nomes_pessoas_lista = sorted(pd.Series(df[cols_autores].values.ravel()).dropna().unique().tolist())

print(f"Total de nomes originais: {len(todos_nomes)}")
print(f"Total de nomes unificados: {len(nomes_pessoas_lista)}")
```

```python
import pandas as pd

# Configuração de colunas conforme o seu ETL
cols_autores = [f'autor_{i}' for i in range(1, 9)]
cols_inst = [f'instituicao_{i}' for i in range(1, 9)]
cols_keywords = [f'keyword_{i}' for i in range(1, 6)]

# --- 1. UNIVERSIDADES (Ranking por presença em artigos) ---
# Conta em quantos artigos únicos cada universidade aparece
df_inst_long = df[['id'] + cols_inst].melt(id_vars='id', value_name='Instituicao').dropna()
ranking_inst = df_inst_long.groupby('Instituicao')['id'].nunique().reset_index(name='Total_Artigos')
ranking_inst = ranking_inst.sort_values('Total_Artigos', ascending=False)
ranking_inst.to_csv('ranking_universidades.csv', index=False, encoding='utf-8-sig')

# --- 2. AUTORES (Filtro: Colaboraram em mais de 1 artigo) ---
# Conta artigos por autor e filtra apenas os que possuem > 1
df_autores_long = df[['id'] + cols_autores].melt(id_vars='id', value_name='Autor').dropna()
ranking_autores = df_autores_long.groupby('Autor')['id'].nunique().reset_index(name='Qtd_Artigos')
ranking_autores = ranking_autores[ranking_autores['Qtd_Artigos'] > 1] # Filtro solicitado
ranking_autores = ranking_autores.sort_values('Qtd_Artigos', ascending=False)
ranking_autores.to_csv('ranking_autores_relevantes.csv', index=False, encoding='utf-8-sig')

# --- 3. UNIVERSIDADE X UNIVERSIDADE (Rede de Colaboração) ---
# Ordenado pelo peso (força da parceria entre as instituições)
rede_inst = df_inst_long.merge(df_inst_long, on='id')
rede_inst = rede_inst[rede_inst['Instituicao_x'] < rede_inst['Instituicao_y']]
colab_inst = rede_inst.groupby(['Instituicao_x', 'Instituicao_y']).size().reset_index(name='Peso_Colaboracao')
colab_inst = colab_inst.sort_values('Peso_Colaboracao', ascending=False)
colab_inst.to_csv('rede_instituicao_x_instituicao.csv', index=False, encoding='utf-8-sig')

# --- 4. AUTOR X AUTOR (Rede de Coautoria) ---
# Ordenado por quem mais trabalha junto
rede_autores = df_autores_long.merge(df_autores_long, on='id')
rede_autores = rede_autores[rede_autores['Autor_x'] < rede_autores['Autor_y']]
colab_autores = rede_autores.groupby(['Autor_x', 'Autor_y']).size().reset_index(name='Trabalhos_Juntos')
colab_autores = colab_autores.sort_values('Trabalhos_Juntos', ascending=False)
colab_autores.to_csv('rede_autor_x_autor.csv', index=False, encoding='utf-8-sig')

# --- 5. PALAVRAS-CHAVE / TEMAS (Ranking de Assuntos) ---
# Ordenado pelas palavras-chave mais frequentes
df_kw_long = df[['id'] + cols_keywords].melt(id_vars='id', value_name='Tema').dropna()
ranking_temas = df_kw_long.groupby('Tema').size().reset_index(name='Frequencia')
ranking_temas = ranking_temas.sort_values('Frequencia', ascending=False)
ranking_temas.to_csv('ranking_temas_artigos.csv', index=False, encoding='utf-8-sig')

print("✅ Todos os 5 CSVs foram gerados e ordenados com sucesso!")
```
